```python
# Detecting Outliers using IQR (Only for numeric columns) 
numeric_cols=df.select_dtypes(include=['number']).columns
Q1=df[numeric_cols].quantile(0.25)
Q3=df[numeric_cols].quantile(0.75)
IQR=Q3-Q1

outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()
print("Outliers Count Per Column:\n", outliers)
```

# Bivariate Analysis

```python
# sale price impact on bedroom
df['BedroomAbvGr'] = df['BedroomAbvGr'].apply(lambda x: 1 if x > 0 else 0 )

plt.figure(figsize=(7,5))
sns.boxplot(x='BedroomAbvGr', y='SalePrice', data=df)
plt.title(' Sale Price impact on BedRoom')
plt.xlabel('BedroomAbvGr')
plt.ylabel(' Sale Price')
plt.tight_layout()
plt.show()
```

```python
# line graph of yearsold vs sale price
yearly_prices = df.groupby('YrSold')['SalePrice'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_prices, x='YrSold', y='SalePrice', marker='o')
plt.title('Average Sale Price by Year Sold')
plt.xlabel('Year Sold')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# boxplot shows the yearsold vs saleprice
df['YearBuiltBucket'] = pd.cut(df['YrSold'], bins=[2005,2006,2007,2008,2009,2010],
                              labels=['<2006','2006-2007','2007-2008','2008-2009','2009-2010'])

plt.figure(figsize=(10,6))
sns.boxplot(x='YearBuiltBucket', y='SalePrice', data=df)
plt.title('Sale Price by Construction Era')
plt.ylabel('Sale Price')
plt.tight_layout()
plt.show()                 
                
```

```python
PropertySize_counts = df['SalePrice'].value_counts().head(20)
plt.figure(figsize=(16, 6))
PropertySize_counts.plot(kind='bar', color= 'blue')
plt.title('Size vs Price')
plt.xlabel('SalePrice')
plt.ylabel('PropertySize')
plt.xticks(rotation=45, ha='right') # Adjust rotation for better readability 
plt.show()
```

```python
# Scatter Plot (PropertySize,PropPrice)
x =df['YrSold']
y =df['SalePrice']

plt.scatter(x,y)
plt.title('Scatter plot of PropertySize vs PropPrice')
plt.xlabel('PropertySize')
plt.ylabel('PropPrice')
plt.show()
```

```python
# scatter plot shows saleprice with yearsold
for feature in year_feature:
    if feature!='YrSold':
        # we will capture the difference between year variable and year the housepricing
        df[feature]=df['YrSold']-df[feature]
        
        data=df.copy()
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
# bargraph shows the number of houses sold per year
sales_volume = df['YrSold'].value_counts().sort_index()
plt.figure(figsize=(10,6))
sns.barplot(x=sales_volume.index, y=sales_volume.values, palette='viridis')
plt.title('Number of Houses Sold per Year')
plt.xlabel('Year')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.show()
```
